# Business Questions

Consultas analíticas sobre el modelo Gold.

```python
from pyspark.sql import functions as F

FACT = "byma.gold.fact_operaciones"
DIM_FECHA = "byma.gold.dim_fecha"
DIM_CLIENTE = "byma.gold.dim_cliente"
DIM_INSTRUMENTO = "byma.gold.dim_instrumento"
DIM_CANAL = "byma.gold.dim_canal"
```

## Volumen operado por instrumento

```python
fact = spark.table(FACT)
dim_instrumento = spark.table(DIM_INSTRUMENTO)

volumen = (
    fact
    .join(dim_instrumento, "instrumento_sk")
    .groupBy("simbolo_titulo")
    .agg(
        F.sum("monto_total").alias("monto_operado"),
        F.count("*").alias("cantidad_operaciones")
    )
    .orderBy(F.desc("monto_operado"))
)

display(volumen.limit(20))
```

## Instrumentos con mayor cantidad de operaciones

```python
top_operaciones = (
    fact
    .join(dim_instrumento, "instrumento_sk")
    .groupBy("simbolo_titulo")
    .count()
    .orderBy(F.desc("count"))
)

display(top_operaciones.limit(20))
```

## Distribución por tipo de operación

```python
tipo_operacion = (
    fact
    .groupBy("tipoTran")
    .count()
    .orderBy(F.desc("count"))
)

display(tipo_operacion)
```

## Volumen operado por fecha

```python
dim_fecha = spark.table(DIM_FECHA)

volumen_fecha = (
    fact
    .join(dim_fecha, "fecha_sk")
    .groupBy("fecha_operacion")
    .agg(
        F.sum("monto_total").alias("monto_operado"),
        F.count("*").alias("cantidad_operaciones")
    )
    .orderBy("fecha_operacion")
)

display(volumen_fecha)
```

## Volumen operado por canal

```python
dim_canal = spark.table(DIM_CANAL)

volumen_canal = (
    fact
    .join(dim_canal, "canal_sk")
    .groupBy("origen")
    .agg(
        F.sum("monto_total").alias("monto_operado"),
        F.count("*").alias("cantidad_operaciones")
    )
    .orderBy(F.desc("monto_operado"))
)

display(volumen_canal)
```
